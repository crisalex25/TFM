###Librerias

```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

###Leer Dataset

```python
AC="https://raw.githubusercontent.com/crisalex25/TFM/main/AC.csv"
GC="https://raw.githubusercontent.com/crisalex25/TFM/main/GC.csv"
PD="https://raw.githubusercontent.com/crisalex25/TFM/main/PrecD.csv"
```

```python
dfAC=pd.read_csv(AC, sep=';', decimal=",")
dfGC=pd.read_csv(GC, sep=';', decimal=",")
dfPD=pd.read_csv(PD, sep=';', decimal=",")
```

###Graficas

```python
df1=dfAC
df2=dfGC
df3=dfPD
```

```python
# Crear un DataFrame de ejemplo
df = pd.DataFrame(df1)

# Convertir la columna 'Fecha_Hora' a formato de fecha y hora
df['Fecha_Hora'] = pd.to_datetime(df['Fecha_Hora'])

# Crear la gráfica de línea de tiempo
plt.plot(df['Fecha_Hora'], df['Total Auto'], marker='o', linestyle='', markersize=0.5)

# Personalizar la gráfica
plt.ylim(bottom=0)
plt.xlabel('Fecha y Hora')
plt.ylabel('Demanda energía horaria Autoconsumos [kWh]')
plt.title('Línea de Tiempo de Energía')
plt.xticks(rotation=45)

# Mostrar la gráfica
plt.show()
```

```python
# Crear un DataFrame de ejemplo
df = pd.DataFrame(df2)

# Convertir la columna 'Fecha_Hora' a formato de fecha y hora
df['Fecha_Hora'] = pd.to_datetime(df['Fecha_Hora'])

# Crear la gráfica de línea de tiempo
plt.plot(df['Fecha_Hora'], df['Total GC'], marker='o', linestyle='', markersize=0.5)

# Personalizar la gráfica
plt.ylim(bottom=0)
plt.xlabel('Fecha y Hora')
plt.ylabel('Demanda energía horaria Gran Consumidor [kWh]')
plt.title('Línea de Tiempo de Energía')
plt.xticks(rotation=45)

# Mostrar la gráfica
plt.show()
```

```python
# Crear un DataFrame de ejemplo
df = pd.DataFrame(df3)

# Convertir la columna 'Fecha_Hora' a formato de fecha y hora
df['Fecha'] = pd.to_datetime(df['Fecha'])

# Crear la gráfica de línea de tiempo
plt.plot(df['Fecha'], df['Valor'], marker='x', linestyle='', markersize=2)

# Personalizar la gráfica
plt.ylim(bottom=0)
plt.xlabel('Fecha y Hora')
plt.ylabel('Precipitación [mm/m2]')
plt.title('Línea de Tiempo Precipitaciones')
plt.xticks(rotation=45)

# Mostrar la gráfica
plt.show()
```

```python
df3 = pd.DataFrame(df3)

# Convertir la columna 'Tiempo' a formato de fecha
df3['Fecha'] = pd.to_datetime(df3['Fecha'])

# Agregar una columna de Meses
df3['Meses'] = df3['Fecha'].dt.month

# Calcular las precipitaciones mensuales
precipitaciones_mensuales = df3.groupby('Meses')['Valor'].mean()

# Crear la gráfica de barras
plt.bar(precipitaciones_mensuales.index, precipitaciones_mensuales)

# Personalizar la gráfica
plt.xlabel('Meses')
plt.ylabel('Precipitaciones (mm/m²)')
plt.title('Precipitaciones Mensuales')
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])

# Mostrar la gráfica
plt.show()
```
